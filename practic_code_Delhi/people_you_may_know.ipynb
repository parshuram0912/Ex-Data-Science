```python
import json   #import the json to deal with the json file
def data_friends(filename):
    with open(filename, "r") as f:
        return json.load(f)

def people_know(data,user_id):
    user_friends={}
    for user in data["users"]:
        user_friends[user["id"]]=set(user["friends"])  #this logic are make the the id of the user friends extract from and data and show the friendts that are set not allow duplicates
    if user_id not in user_friends:
        return []
    

data=data_friends("student_info.json")
user_id=1
people_know(data,1)

        
```
