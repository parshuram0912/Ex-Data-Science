```python
import numpy as np
myarr=np.array([[1,2,3],[4,5,6]])
print(myarr)
```

```python
flat=np.array(myarr.flatten())
print(flat)  #this function will convert the 2d array into 1d array
```

```python
flat[0:3]
```

```python
new_array=flat[2:5] #this is slicing the array and creating a new array from the original array
print(new_array)
```
