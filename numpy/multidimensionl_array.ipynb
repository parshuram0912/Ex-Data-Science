indexing and multidimenssion array


```python
arr2=np.array([[6,8,9,],[7,3,8],[7,3,4,]])
print(arr2)
print(arr2[0,1])  #it shows the element on rows and column 1 is that 8
```

```python
print(arr2[0:2, 1:3])  # Extracts first 2 rows and last 2 columns
```
