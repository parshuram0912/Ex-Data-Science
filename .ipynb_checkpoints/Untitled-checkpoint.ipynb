```python
import json
data = {
  "library_name": "City Central Library",
  "location": "Mumbai",
  "active_members": 150,
  "books": [
    {
      "id": 101,
      "title": "The Great Gatsby",
      "author": "F. Scott Fitzgerald",
      "available": true
    },
    {
      "id": 102,
      "title": "1984",
      "author": "George Orwell",
      "available": false
    },
    {
      "id": 103,
      "title": "Pride and Prejudice",
      "author": "Jane Austen",
      "available": true
    }
  ]
}
python_conversion=json.dumps(data)
print(python_conversion)

type(python_conversion)
```
